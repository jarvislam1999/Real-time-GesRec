```python
import cv2
import numpy as np
import os
from os.path import isfile, join

from ipywidgets import Video, Image
from IPython.display import display
import base64
```

```python
path_to_image = '/mnt/data/jarvislam1999/data/tmp/subject01_setting1_03_syn_bg0001/'
gesture_order_path = '/mnt/data/yxchen/gesture-datasets/ems/data/subject01_machine_recovery_3gps_2pairs_04/'
```

```python
f = open(gesture_order_path+'quick.txt', 'r')
image_folders = []
```

```python
pron_count = 1
susp_count = 1
left_count = 1
righ_count = 1
for line in f:
    if "Pronation" in line and "start" in line:
        image_folders.append('quick_pronation_%s%d_all' % ("0" if pron_count <= 9 else "", pron_count))
        pron_count += 1
    if "Supination" in line and "start" in line:
        image_folders.append('quick_supination_%s%d_all' % ("0" if susp_count <= 9 else "", susp_count))
        susp_count += 1
    if "Left" in line and "start" in line:
        image_folders.append('quick_wrist_left_%s%d_all' % ("0" if left_count <= 9 else "", left_count))
        left_count += 1
    if "Right" in line and "start" in line:
        image_folders.append('quick_wrist_right_%s%d_all' % ("0" if righ_count <= 9 else "", righ_count))
        righ_count += 1
```

```python
image_list = []
```

```python
for folder in image_folders:
    if not os.path.exists(path_to_image+'rgb/'+folder):
        continue
    images = os.listdir(path_to_image+'rgb/'+folder)
    images.sort()
    for image in images:
        image_list.append(path_to_image+'rgb/'+folder+'/'+image)
```

```python
len(image_list)
```

```python
frame_array = []
```

```python
for filename in image_list:
    #reading each files
    img = cv2.imread(filename)
    height, width, layers = img.shape
    size = (width,height)
    
    #inserting the frames into an image array
    frame_array.append(img)
```

```python
fps = 30
out = cv2.VideoWriter(path_to_image+'video.mp4',cv2.VideoWriter_fourcc(*'MP4V'), fps, size)
```

```python
for i in range(len(frame_array)):
    # writing to a image array
    out.write(frame_array[i])
out.release()
```
