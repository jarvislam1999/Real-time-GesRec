```python
print(cv2.__version__)
vidcap = cv2.VideoCapture('/mnt/data/jarvislam1999/data/subject01_bg/still_crerar.mp4')
success,image = vidcap.read()
count = 0
success = True
while success:
    name = ("0000%d.jpg" % count if (count <= 9) \
            else ("000%d.jpg" % count if (count <=99) \
                  else ("00%d.jpg" %count if (count <= 999) else ("0%d.jpg" %count) )) )
    
    cv2.imwrite("/mnt/data/jarvislam1999/data/subject01_bg/still_crerar/" + name, image)     # save frame as JPEG file
    success,image = vidcap.read()
    print('Read a new frame: ', success)
    count += 1
```
